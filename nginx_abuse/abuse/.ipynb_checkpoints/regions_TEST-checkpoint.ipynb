```python
%%time
reg_id = [i for i in regions_dic.values()]
total_data = pd.DataFrame()
for i in range(450):
    chunk_name = f'chunk{i}.csv'
    chunk = pd.read_csv(chunk_name)
    new_data = chunk[chunk['region_id'].isin(reg_id)].copy()
    if new_data.shape[0] != 0:
        total_data = total_data.append(new_data)
```

```python
all_cities = total_data.copy()
```

```python
df = all_cities[['region_ru', 'area_ru', 'title_ru']]
```

```python
mi = df[df['region_ru'] == 'Минская область']
```

```python
# регион>область
dic = {}
for reg in regs:
    this_reg = df[df['region_ru'] == reg]
    dic[reg] = this_reg['area_ru'].unique()

    
```

```python
%%time
dic = {}
area_dic = {}
city_dic = {}
for reg in regs:
    this_reg = df[df['region_ru'] == reg]
    dic[reg] = {}
    for area in this_reg['area_ru']:
        dic[reg][area] = []
        for city in df[df['area_ru'] == area]['title_ru']:
             dic[reg][area].append(city)
        
        

        
        
    
    
```
