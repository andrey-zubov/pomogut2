```python
import codecs
with codecs.open(open('test_org.txt', 'r') as file:
    text = file.readlines()
```

```python
text
```

```python
buffer = []
orgs = []
for line in text:
    if line[:3] != '---':
        s = line.split(':', 1)
        buffer.append(s)
    else:
        orgs.append(buffer)
        buffer = []
```
